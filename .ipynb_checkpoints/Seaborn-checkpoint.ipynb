```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins['species'].value_counts()
```

```python
penguins['island'].value_counts()
```

```python
# Scatterplot
sns.set_style("ticks")
sns.set_context("notebook")
sns.scatterplot(data = penguins,x = "flipper_length_mm", y = "body_mass_g", hue = "island",palette = "Dark2")
sns.despine()
```

```python
sns.scatterplot(data = penguins, x = "species",y = "body_mass_g",hue="island",style = "sex")
```

```python
sns.stripplot(data = penguins, x="species",y="body_mass_g",hue="island",dodge = "True")
```

```python
# Swarmplot
sns.set_context("talk")
sns.swarmplot(data = penguins, x="species",y="body_mass_g",hue="island")
sns.despine()
```

```python
# Histogram
sns.set_context("notebook")
sns.histplot(data = penguins, x = "body_mass_g", hue = "sex", multiple = "stack")
```

```python
# Regression plot
sns.regplot(data = penguins, x="body_mass_g", y = "flipper_length_mm",scatter=False)
# scatter = false makes the data points invisible
```

```python
# Line Plot
sns.set_style("whitegrid")
sns.lineplot(data = penguins, x="body_mass_g", y = "flipper_length_mm",hue="sex")
```

```python
# Joint Plot

sns.jointplot(data = penguins, x="body_mass_g", y = "flipper_length_mm")
```

```python
sns.jointplot(data = penguins, x="body_mass_g", y = "flipper_length_mm",hue="sex")
```

```python
#bar plot

sns.barplot(data = penguins, x="species", y = "flipper_length_mm",hue = "sex",palette=["skyblue","pink"])
```

```python
# count plot

sns.countplot(data = penguins, x="species",hue = "island")
```
