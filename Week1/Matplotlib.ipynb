### Loading Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Real World Examples

#### Line Graph

```python
gas = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/refs/heads/master/gas_prices.csv')

plt.title('Gas Prices over Time (in USD)',fontdict={'fontweight':'bold','fontsize':16})

plt.plot(gas.Year,gas.USA,'b.-', label = 'United States')
plt.plot(gas.Year,gas.Canada,'g.-', label = 'Canada')
plt.plot(gas['Year'],gas['South Korea'],'y.-', label = 'South Korea')

print(gas.Year[::3])

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/refs/heads/master/fifa_data.csv')

fifa.head(5)
```

### Histograms

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall,bins=bins,color='green')

plt.xticks(bins)
plt.ylabel('Number of players')
plt.xlabel('Skill level')
plt.title('Distribution of player skills in Fifa')

plt.show()
```

### Pie Charts

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].shape[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].shape[0]

labels = ['Left','Right']
colors = ['red','blue']

plt.pie([left,right], labels = labels, colors = colors, autopct = '%.2f %%')

plt.title('Foot Preference Of FIFA Players')
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

plt.style.use('ggplot')
light = fifa.loc[fifa.Weight < 125].shape[0]
light_med = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150) ].shape[0]
med = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175) ].shape[0]
med_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200) ].shape[0]
heavy = fifa.loc[(fifa.Weight >= 200) ].shape[0]

weights = [light,light_med,med,med_heavy,heavy]
labels = ['Under 125','125-150','150-175','175-200','Over 200']

plt.title('Weight distribution of FIFA Players')

plt.pie(weights,labels = labels, autopct = '%.2f %%',pctdistance = 0.8)

plt.show()




```

```python
plt.style.use('default')

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['Barca','Madrid','Revs']

boxes = plt.boxplot([barcelona,madrid,revs],labels = labels ,patch_artist = True,medianprops={'linewidth':2})

for box in boxes['boxes']:
    # Set edge color
    box.set(color='#4286f4',linewidth=2)

    #Change fill color
    box.set(facecolor='#e0e0e0')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
    
```
