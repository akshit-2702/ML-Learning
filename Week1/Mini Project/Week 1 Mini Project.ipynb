## Iris Dataset Analysis

```python
# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Basic Exploration

```python
# Set Seaborn style
sns.set(style="whitegrid")

#Load Dataset
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
print("First 5 rows of the dataset:")
display(df.head())

# 3. Basic Info
print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())
```

### Visualizations using different plots

```python
# 4. Pairplot
sns.pairplot(df, hue="species", diag_kind="kde", height=2)
plt.suptitle("Pairwise Relationships of Iris Features", y=1.02, fontsize=14)
plt.show()

# 5. Scatter Plot
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="sepal_length", y="petal_length", hue="species", s=70, palette="deep")
plt.title("Sepal Length vs Petal Length")
plt.show()

# 6. Correlation Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df.drop(columns="species").corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()

# 7. Boxplots
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="species", y="petal_length", palette="Set2")
plt.title("Petal Length Distribution by Species")
plt.show()
```
